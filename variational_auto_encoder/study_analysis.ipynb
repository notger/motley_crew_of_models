```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('vae_study.csv', ignore)
```

```python
df.head()
```

```python
list(df.test_losses)
```

```python

fig = plt.figure(figsize=(8, 6))  # Create a figure instance
ax = fig.add_subplot(111)  # Add axes to the figure
df.sort_values('k').groupby('n_embeddings').plot(x='k', y='test_losses', kind='line', sharex=True, sharey=True, ax=ax)
plt.legend(list(df.n_embeddings.unique()))
```
