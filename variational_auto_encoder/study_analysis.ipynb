```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('vae_study_n_informative_5.csv')
```

```python
df.head()
```

```python
fig = plt.figure(figsize=(8, 6))  # Create a figure instance
ax = fig.add_subplot(111)  # Add axes to the figure
df.sort_values('k').groupby('n_embeddings').plot(x='k', y='test_losses', kind='line', sharex=True, sharey=True, ax=ax, title='Test losses')
plt.legend(list(df.n_embeddings.unique()))
```
