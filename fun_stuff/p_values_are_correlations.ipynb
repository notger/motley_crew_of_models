There is this famous claim, that "p-values are just correlation values in a trenchcoat", and we want to look at that.

```python
import numpy as np
import scipy.stats
import matplotlib.pyplot as plt

x, y = [], []

for _ in range(100):
    N = 1_000
    treatment = np.random.randint(0, 2, N)
    outcome = np.random.randint(0, 2, N)
    score = np.corrcoef(treatment, outcome)[0, 1]*np.sqrt(N)
    y.append(scipy.stats.norm.sf(abs(score)) * 2)
    x.append(scipy.stats.ttest_ind(a=treatment, b=outcome, equal_var=True).pvalue)

plt.scatter(x, y)
plt.title('p_value calculated for two random binary arrays of length 1_000')
plt.xlabel('p_value classic')
plt.ylabel('p_value via correlation + z_score')
```

Nope the p-value is not just a correlation in a trenchcoat.
